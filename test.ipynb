```python
# Install dependencies (skip if already installed)
# !pip install transformers torch scikit-learn -q

from transformers import AutoTokenizer, AutoModel
import torch
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load SPECTER model
tokenizer = AutoTokenizer.from_pretrained("allenai/specter")
model = AutoModel.from_pretrained("allenai/specter")

def get_embedding(text):
    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        padding=True,
        max_length=512
    )
    
    with torch.no_grad():
        output = model(**inputs)
    
    # CLS token embedding
    embedding = output.last_hidden_state[:,0,:]
    
    return embedding.squeeze().numpy()


# Paragraph 1: Computer networks leading to cybersecurity
para1 = """
Computer networks enable communication between multiple systems across the internet.
As networks grow in size and complexity, they become vulnerable to attacks such as
unauthorized access, data interception, and malware propagation. These risks have
led to the development of cybersecurity practices that protect network infrastructure,
ensure confidentiality of data, and maintain system integrity.
"""

# para1 = """
# Tennis is a sport played with racket and balls. It involves 2 or 4 players on the court.
# It is called as the gentlemen's sport. It is one of the hardest and most expensive sports
# to reach the pinnacle. Roger Federer redefined the elegance, Rafa showed the brutality
# and Novak showed the precision required to play the sport.
# """

# Paragraph 2: Cybersecurity
para2 = """
Cybersecurity focuses on protecting computer systems, networks, and digital data
from malicious attacks. It involves techniques such as encryption, authentication,
and intrusion detection systems to prevent unauthorized access and maintain the
confidentiality, integrity, and availability of information.
"""


# Generate embeddings
emb1 = get_embedding(para1)
emb2 = get_embedding(para2)

# Compute cosine similarity
similarity = cosine_similarity([emb1], [emb2])[0][0]

# Compute cosine distance
distance = 1 - similarity

print("Embedding dimension:", len(emb1))
print("Cosine Similarity:", similarity)
print("Cosine Distance:", distance)
```
